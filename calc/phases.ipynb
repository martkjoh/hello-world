```python
from IPython.display import display, Latex

def pprint(expr):
    return display(Latex("$$" + latex(expr) +"$$"))
```

$$
    -2 \epsilon \sin(\chi) \cos(\chi) e^{i (\varphi_u + \varphi_v)} = |\Delta| [ \cos(\chi)^2 e^{i(2 \varphi_u + \varphi)} + \sin(\chi)^2 e^{i(2 \varphi_v - \varphi)} ]
$$

$$
    -2 \epsilon \sin(\chi) \cos(\chi) = |\Delta| [ \cos(\chi)^2 e^{i( \varphi_u + \varphi - \varphi_v)} + \sin(\chi)^2 e^{i(\varphi_v - \varphi - \varphi_u)} ]
$$
La $y = \varphi_u + \varphi - \varphi_v$
$$
    -2 \epsilon \sin(\chi) \cos(\chi) = |\Delta| [ \cos(\chi)^2 + \sin(\chi)^2 e^{-i2y} ] e^{iy}
$$


```python
f(x, y) = (cos(x)**2 + sin(x)**2 * exp(-i*2*y))*exp(i*y)
pprint(f)
```
