```python
f(x, y, t) = -2 * (cos(x) + cos(y) + 2*t * cos(x)*cos(y))
pprint(f)
```

```python
contour_plot(f(x, y, 0), (x, -pi, pi), (y, -pi, pi), fill=False, labels=True)
```

```python
contour_plot(f(x, y, 0.1), (x, -pi, pi), (y, -pi, pi), fill=False, labels=True)
```

```python
contour_plot(f(x, y, -0.1), (x, -pi, pi), (y, -pi, pi), fill=False, labels=True, contours=[0])
```
